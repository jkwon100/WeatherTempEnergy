```python
# Cell: Analyze Cleaned Energy Data
import pandas as pd
import os

energy_dir = "energyData_clean"
file_list = [f for f in os.listdir(energy_dir) if f.endswith("_hourly_daily.csv")]

summary = []
total_rows = 0

for filename in file_list:
    path = os.path.join(energy_dir, filename)
    df = pd.read_csv(path, index_col=0, parse_dates=True)

    region = filename.split("_")[0]
    row_count = len(df)
    total_rows += row_count

    summary.append({
        "Region": region,
        "Entries": row_count
    })

# Display region-level and total summary
summary_df = pd.DataFrame(summary).sort_values("Region").reset_index(drop=True)
print(f"\nðŸ“Š Total combined entries: {total_rows}")
summary_df
```
